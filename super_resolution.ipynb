```python
image = cv2.imread('/content/1_A8yToxEh-f0_1Up8u51aHQ.png')
path = "/content/EDSR_x4.pb"
sr.readModel(path)
sr.setModel("edsr", 4)
result = sr.upsample(image)
cv2.imwrite('/content/upscaled1.jpeg', result)
```
